```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import joblib
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
file = os.path.join('C:/Users/jorge/Downloads', 'Chancador_Raw_Data.csv')
df1 = pd.read_csv(file)
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
df1 = df1.drop([
	'Corriente.1',
	'Corriente.2',
	'Pesómetro',
	'V. Cero',
	'Corriente.3',
	'V. Cero.1',
	'Corriente.4',
	'V. Cero.2',
	'Unnamed: 0',
], axis='columns')
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
df1 = df1.rename(columns={
	'Salto Anillo': 'Salto Anillo A',
	'Unnamed: 9': 'Salto Anillo B',
	'Unnamed: 10': 'Salto Anillo C',
	'Unnamed: 11': 'Salto Anillo D',
})
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
df1 = df1
s = df1['Tiempo']
df1['Tiempo'] = pd.to_datetime(s, errors='coerce')

s = df1['Corriente']
df1['Corriente'] = pd.to_numeric(s, errors='coerce')

s = df1['Presión C. Eje']
df1['Presión C. Eje'] = pd.to_numeric(s, errors='coerce')

s = df1['T° Retorno']
df1['T° Retorno'] = pd.to_numeric(s, errors='coerce')

s = df1['T° socket Linner']
df1['T° socket Linner'] = pd.to_numeric(s, errors='coerce')

s = df1['T° Excéntrica']
df1['T° Excéntrica'] = pd.to_numeric(s, errors='coerce')

s = df1['T° Alimentación']
df1['T° Alimentación'] = pd.to_numeric(s, errors='coerce')

s = df1['T° Desc. C. Eje']
df1['T° Desc. C. Eje'] = pd.to_numeric(s, errors='coerce')

s = df1['Salto Anillo A']
df1['Salto Anillo A'] = pd.to_numeric(s, errors='coerce')

s = df1['Salto Anillo B']
df1['Salto Anillo B'] = pd.to_numeric(s, errors='coerce')

s = df1['Salto Anillo C']
df1['Salto Anillo C'] = pd.to_numeric(s, errors='coerce')

s = df1['Salto Anillo D']
df1['Salto Anillo D'] = pd.to_numeric(s, errors='coerce')

s = df1['Nivel Taza']
df1['Nivel Taza'] = pd.to_numeric(s, errors='coerce')

s = df1['Setting']
df1['Setting'] = pd.to_numeric(s, errors='coerce')

attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
df1 = df1.dropna(axis='index')

attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
df1 = df1
df1.set_index([
	'Tiempo',
], inplace=True)
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
from functools import partial
from pandas.tseries.frequencies import to_offset

def roundfunc(t, freq):
	freq = to_offset(freq)
	return pd.Timestamp((t.value // freq.delta.value) * freq.delta.value)

df18 = df1.groupby(partial(roundfunc, freq='4T'))

attrs = {
	'var': 'df18',
	'type': type(df18).__name__,
}
[attrs]
df18
enum = enumerate(df18.groups)
nextel = next(enum)
nextel[1]
df18.get_group(nextel[1])
nextel = next(enum)
nextel[1]
df18.get_group(nextel[1])
df18 = df18.aggregate('median')

attrs = {
	'var': 'df18',
	'type': type(df18).__name__,
}
[attrs]
df18
file = os.path.join('C:/Users/jorge/OneDrive/Desktop/ING. MANT. INDUSTRIAL/Computación Aplicada', 'Chancador4min.csv')
df18.to_csv(file)
attrs = {
	'var': 'empty',
	'type': 'empty',
}
[attrs]
'csv file saved'
from functools import partial
from pandas.tseries.frequencies import to_offset

def roundfunc(t, freq):
	freq = to_offset(freq)
	return pd.Timestamp((t.value // freq.delta.value) * freq.delta.value)

df20 = df1.groupby(partial(roundfunc, freq='8T'))

attrs = {
	'var': 'df20',
	'type': type(df20).__name__,
}
[attrs]
df20
enum = enumerate(df20.groups)
nextel = next(enum)
nextel[1]
df20.get_group(nextel[1])
nextel = next(enum)
nextel[1]
df20.get_group(nextel[1])
df20 = df20.aggregate('median')

attrs = {
	'var': 'df20',
	'type': type(df20).__name__,
}
[attrs]
df20
file = os.path.join('C:/Users/jorge/OneDrive/Desktop/ING. MANT. INDUSTRIAL/Computación Aplicada', 'CHANCADOR8MIN.csv')
df20.to_csv(file)
attrs = {
	'var': 'empty',
	'type': 'empty',
}
[attrs]
'csv file saved'
from functools import partial
from pandas.tseries.frequencies import to_offset

def roundfunc(t, freq):
	freq = to_offset(freq)
	return pd.Timestamp((t.value // freq.delta.value) * freq.delta.value)

df23 = df1.groupby(partial(roundfunc, freq='12T'))

attrs = {
	'var': 'df23',
	'type': type(df23).__name__,
}
[attrs]
df23
enum = enumerate(df23.groups)
nextel = next(enum)
nextel[1]
df23.get_group(nextel[1])
nextel = next(enum)
nextel[1]
df23.get_group(nextel[1])
df23 = df23.aggregate('median')

attrs = {
	'var': 'df23',
	'type': type(df23).__name__,
}
[attrs]
df23
file = os.path.join('C:/Users/jorge/OneDrive/Desktop/ING. MANT. INDUSTRIAL/Computación Aplicada', 'chancador12min.csv')
df23.to_csv(file)
attrs = {
	'var': 'empty',
	'type': 'empty',
}
[attrs]
'csv file saved'
file = os.path.join('C:/Users/jorge/OneDrive/Desktop/ING. MANT. INDUSTRIAL/Computación Aplicada/', 'DetencionesRev.xlsx')
df9 = pd.read_excel(file, header=0)
attrs = {
	'var': 'df9',
	'type': type(df9).__name__,
}
[attrs]
df9
df9 = df9.drop([
	'Unnamed: 0',
	'Día',
	'Fecha',
	'Detalle de Reporte',
	'Tiempo Detención (hrs)',
	'Tipo de Detención (2)',
	'Año',
	'MES',
	'Estado',
], axis='columns')
attrs = {
	'var': 'df9',
	'type': type(df9).__name__,
}
[attrs]
df9
cond = ((df9['Equipo'].eq('130-CV-004')))
df9 = df9[cond]
attrs = {
	'var': 'df9',
	'type': type(df9).__name__,
}
[attrs]
df9
cond = ((df9['Tipo de Detención'].isin(["Operacional", "Mantención Programada"])))
df9['Tipo de Detención'].mask(cond, other='Pausa', inplace=True)

attrs = {
	'var': 'df9',
	'type': type(df9).__name__,
}
[attrs]
df9
file = os.path.join('C:/Users/jorge/OneDrive/Desktop/ING. MANT. INDUSTRIAL/Computación Aplicada', 'Chancador4min.csv')
df8 = pd.read_csv(file, header=0)
attrs = {
	'var': 'df8',
	'type': type(df8).__name__,
}
[attrs]
df8
df8 = df8
s = df8['Tiempo']
df8['Tiempo'] = pd.to_datetime(s, errors='coerce')

attrs = {
	'var': 'df8',
	'type': type(df8).__name__,
}
[attrs]
df8
df8 = df8
df8.set_index([
	'Tiempo',
], inplace=True)
attrs = {
	'var': 'df8',
	'type': type(df8).__name__,
}
[attrs]
df8
col_start = df9['Inicio']
col_end = df9['Fin']
idx = pd.IntervalIndex.from_arrays(col_start, col_end, closed='left')
df9_interval = df9.set_index(idx)
df9_interval = df9_interval.drop(columns=[
	'Inicio',
	'Fin'
], axis='columns')
df_labels15 = pd.DataFrame(index=df8.index)
df_labels15['Estado de Salud'] = 'Operativo'
for i, r in df9_interval.iterrows():
	mask = (df_labels15.index >= i.left) & (df_labels15.index < i.right)
	df_labels15.loc[mask, 'Estado de Salud'] = r['Tipo de Detención']
attrs = {
	'var': 'df_labels15',
	'type': type(df_labels15).__name__,
}
[attrs]
df_labels15
df16 = pd.concat([
	df8,
	df_labels15,
], axis='columns', join='outer')

attrs = {
	'var': 'df16',
	'type': type(df16).__name__,
}
[attrs]
df16
cond = ((df16['Estado de Salud'].ne('Pausa')))
df16 = df16[cond]
attrs = {
	'var': 'df16',
	'type': type(df16).__name__,
}
[attrs]
df16
df16 = df16
cond = ((df16['Estado de Salud'].eq('Operativo')))
df16['Estado de Salud'].mask(cond, other=0, inplace=True)

cond = ((df16['Estado de Salud'].eq('Falla')))
df16['Estado de Salud'].mask(cond, other=1, inplace=True)

attrs = {
	'var': 'df16',
	'type': type(df16).__name__,
}
[attrs]
df16
df26 = df16.drop([
	'Estado de Salud',
], axis='columns')
attrs = {
	'var': 'df26',
	'type': type(df26).__name__,
}
[attrs]
df26
df27 = df16[[
	'Estado de Salud',
]]
attrs = {
	'var': 'df27',
	'type': type(df27).__name__,
}
[attrs]
df27
def datetime_window(dfx, dfy, dfy_total, scaler, size, periods, y_choose='last', ahead=''):
    catcolsx = dfx.select_dtypes(['category']).columns
    dfx[catcolsx] = dfx[catcolsx].apply(lambda x: x.cat.codes)

    catcolsy = dfy.select_dtypes(['category']).columns
    dfy[catcolsy] = dfy[catcolsy].apply(lambda x: x.cat.codes)

    catcolsy_total = dfy_total.select_dtypes(['category']).columns
    dfy_total[catcolsy_total] = dfy_total[catcolsy_total].apply(lambda x: x.cat.codes)

    ind = []
    x = []
    y = []

    def get_windows(win):
        if (win.shape[0] != periods):
            return np.nan
        
        if (y_choose == 'ahead'):
            wy = dfy.loc[win.index, :]
            try:
                wy = dfy_total.loc[wy.index[-1] + pd.Timedelta(ahead)]
                y.append(wy)

                wx = dfx.loc[win.index, :]
                x.append(wx)

                ind.append(win.index[0])
            except:
                pass
        else:
            wx = dfx.loc[win.index, :]
            x.append(wx)

            wy = dfy.loc[win.index, :]
            if (y_choose == 'last'):
                wy = wy.iloc[-1]
            elif (y_choose == 'first'):
                wy = wy.iloc[0]
            y.append(wy)

            ind.append(win.index[0])
        
        return np.nan

    dfx.iloc[:, 0].rolling(size).apply(get_windows)
    
    return x, y, ind

# --------------------------------------------------------------------------------------
def print_list_shape(lst):
    s = ''
    if (len(lst) > 0):
        s += str(len(lst))
        s += ' x ' + str(np.array(lst[0]).shape)
    else:
        s += 'empty'
        
    return s

# --------------------------------------------------------------------------------------
def normal_anomal_split(dfx, dfy, size, periods, y_choose='last', ahead='', custom_normal_split=False, normal_test_ratio=0.1, scaler='MinMaxScaler'):
    
    if (scaler == 'MinMaxScaler'):
        myscaler = MinMaxScaler(feature_range=(0, 1))
    elif (scaler == 'StandardScaler'):
        myscaler = StandardScaler()
    elif(scaler == 'RobustScaler'):
        myscaler = RobustScaler(with_centering=True, with_scaling=True, quantile_range=(25, 75))
    
    x_windowed, y_windowed, windows_ind = datetime_window(dfx=dfx, dfy=dfy, dfy_total=dfy, scaler=myscaler, size=size, periods=periods, y_choose=y_choose, ahead=ahead)
    
    
    x_normal = []
    y_normal = []
    x_anomal = []
    y_anomal = []
    for (i, wy) in enumerate(y_windowed):
        if wy.iloc[0] == 0:
            x_normal.append(x_windowed[i])
            y_normal.append(y_windowed[i])
        else:
            x_anomal.append(x_windowed[i])
            y_anomal.append(y_windowed[i])
    
    
    normal_indices = [i for i in range(len(x_normal))]
    normal_indices_permuted = np.random.permutation(normal_indices)
    
    length = len(x_normal)
    length_test = round(length * normal_test_ratio)
    
    if custom_normal_split:
        normal_test_indices = normal_indices_permuted[:length_test]
        normal_train_indices = normal_indices_permuted[length_test:]
    else:
        normal_test_indices = normal_indices_permuted[:len(y_anomal)]
        normal_train_indices = normal_indices_permuted[len(y_anomal):]
    
    
    x_normal_train = [x_normal[i] for i in normal_train_indices]
    x_normal_test = [x_normal[i] for i in normal_test_indices]
    
    
    x_normal_train_total = pd.concat(x_normal_train)
    x_normal_train_total = x_normal_train_total[~x_normal_train_total.index.duplicated(keep='first')]
    x_normal_train_total.sort_index(inplace=True)
    
    myscaler.fit(x_normal_train_total.to_numpy())
    
    x_train = []
    for wx in x_normal_train:
        wx = wx.to_numpy()
        wx = np.squeeze(myscaler.transform(wx))
        x_train.append(wx)
    
    
    x_test = []
    y_test = []
    for wx in x_normal_test:
        wx = wx.to_numpy()
        wx = np.squeeze(myscaler.transform(wx))
        x_test.append(wx)
        y_test.append(0)
        
    for wx in x_anomal:
        wx = wx.to_numpy()
        wx = np.squeeze(myscaler.transform(wx))
        x_test.append(wx)
        y_test.append(1)
    
    x_mix = []
    for wx in x_windowed:
        wx = wx.to_numpy()
        wx = np.squeeze(myscaler.transform(wx))
        x_mix.append(wx)
    
    y_mix = []
    for wy in y_windowed:
        wy = wy.to_numpy()
        wy = wy[0]
        y_mix.append(wy)
        
    i_mix = windows_ind

    
    return x_train, x_test, y_test, x_mix, y_mix, i_mix, myscaler
x_train, x_test, y_test, x_mix, y_mix, i_mix, myscaler = normal_anomal_split(
	dfx=df26, dfy=df27
	, size='20T', periods=5, y_choose='last', ahead=''
	, custom_normal_split=True, normal_test_ratio=0.2, scaler='MinMaxScaler')

df26 = {
	'x_train': np.array(x_train),
	'x_test': np.array(x_test),
	'y_test': np.array(y_test),
	'x_mix': np.array(x_mix),
	'y_mix': np.array(y_mix),
	'i_mix': i_mix,
	'myscaler': myscaler,
}

attrs = {
	'var': 'df26',
	'type': type(df26).__name__,
}
[attrs]

{
	'x_train': print_list_shape(x_train),
	'x_test': print_list_shape(x_test),
	'y_test': print_list_shape(y_test),
}

np.squeeze(df26['x_train'])
np.squeeze(df26['x_test'])
np.squeeze(df26['y_test'])
file = os.path.join('C:/Users/jorge/OneDrive/Desktop/ING. MANT. INDUSTRIAL/Computación Aplicada', 'Ventanas4T.NPZ')
np.savez(file, x_train=df26['x_train'], x_test=df26['x_test'], y_test=df26['y_test'], x_mix=df26['x_mix'], y_mix=df26['y_mix'], i_mix=df26['i_mix'])
scaler_file = os.path.join('C:/Users/jorge/OneDrive/Desktop/ING. MANT. INDUSTRIAL/Computación Aplicada', 'scaler.joblib')
joblib.dump(df26['myscaler'], scaler_file)
attrs = {
	'var': 'empty',
	'type': 'empty',
}
[attrs]
'npz file saved'
```
